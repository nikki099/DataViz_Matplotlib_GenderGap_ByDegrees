```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']
```

```python
year=women_degrees['Year']
stem=women_degrees[stem_cats]
arts=women_degrees[lib_arts_cats]
other=women_degrees[other_cats]
```

```python
fig=plt.figure(figsize=(18,20))
 
#1st col = stem
for x in range(6):#[1,4,7,10,13,16]
    sp=x*3+1
    ax=fig.add_subplot(6,3,sp)
    ax.plot(year,stem.iloc[:,x],c=cb_dark_blue, label='women',linewidth=3)
    ax.plot(year,100-(stem.iloc[:,x]),c=cb_orange, label='men',linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[x])
    #remove tick params
    ax.tick_params(top='off', bottom='off', right='off',left='off')
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    if sp==1:
        ax.text(2005,85,"Women")
        ax.text(2005,30,"Men")
    elif sp==16:
        ax.text(2005,90,"Men")
        ax.text(2005,7,"Women")

#2nd col = arts

for x in range(5): #[2,5,8,11,14]
    sp=x*3+2
    ax=fig.add_subplot(6,3,sp)
    ax.plot(year,arts.iloc[:,x],c=cb_dark_blue, label='women',linewidth=3)
    ax.plot(year,100-(arts.iloc[:,x]),c=cb_orange, label='men',linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[x])
    #remove tick params
    ax.tick_params(top='off', bottom='off', right='off',left='off')
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    if sp==2:
        ax.text(2005,80,"Women")
        ax.text(2005,20,"Men")


#3rd col = other

for x in range(6): #[3,6,9,12,15,18]
    sp=(x+1)*3
    ax=fig.add_subplot(6,3,sp)
    ax.plot(year,other.iloc[:,x],c=cb_dark_blue, label='women',linewidth=3)
    ax.plot(year,100-(other.iloc[:,x]),c=cb_orange, label='men',linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[x])
    #remove tick params
    ax.tick_params(top='off', bottom='off', right='off',left='off')
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    if sp==3:
        ax.text(2005,90,"Women")
        ax.text(2005,8,"Men")  
        
    elif sp==18:
        ax.text(2005,80,"Men")
        ax.text(2005,20,"Women") 

   
    
plt.show()

  
```
