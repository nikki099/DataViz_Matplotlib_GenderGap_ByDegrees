```python
#chapter 2 df and scatter plot
```

```python
#Scatterplot - Sample_size and Median

ax=recent_grads.plot(x='Sample_size', y='Median', kind='scatter', 
                  title='Median vs. Sample_size', figsize=(5,10))

ax.set_xlim(0,5000)
```

```python
#Scatterplot - Sample_size and Unemployment_Rate

ax=recent_grads.plot(x='Sample_size', y='Unemployment_rate',
                  kind='scatter', title='Sample_size vs Unemployment_Rate',
                  figsize=(5,10))

ax.set_xlim(0,5000)
```

```python
#Scatterplot -Fulltime and Median

ax=recent_grads.plot(x='Full_time', y='Median',
                  kind='scatter', title='Fulltime and Median',
                  figsize=(5,10))

ax.set_xlim(0,300000)
```

```python
#Scatterplot - ShareWomen and Unemployment_Rate

ax=recent_grads.plot(x='ShareWomen', y='Unemployment_rate',
                  kind='scatter', title='Sample_size vs Unemployment_Rate',
                  figsize=(5,10))

ax.set_xlim(0,1.2)
```

```python
#Scatterplot - Men vs Median

ax=recent_grads.plot(x='Men', y='Median',
                  kind='scatter', title='Men vs Median',
                  figsize=(5,10))

ax.set_xlim(0,200000)
```

```python
#Scatterplot - Women vs Median

ax=recent_grads.plot(x='Women', y='Median',
                  kind='scatter', title='Men vs Median',
                  figsize=(5,10))

ax.set_xlim(0,200000)
```

```python
#Do students in more popular majors make more money?
#Answer - yes

ax=recent_grads.plot(x='Rank', y='Median', kind='Scatter', figsize=(5,10))
ax.set_title("Do students in more popular majors make more money?")
ax.set_xlim(0,200)
```

```python
#Do students that majored in subjects that were majority female make more money?
#Answer yes

ax=recent_grads.plot(x='ShareWomen', y='Median',
                     title="Do students majored in subs that were majority women make more money",
                     kind='scatter',
                     figsize=(5,10))
ax.set_xlim(0,1.2)
```

```python
#Is there any link between the number of full-time employees and median salary?
#Yes, but weak.

ax=recent_grads.plot(x='Full_time', y='Median',
                     kind='scatter',
                     title="Fulltime vs Meidan Salary",
                     figsize=(5,10)                                 
                    )

ax.set_xlim(0,300000)
```
